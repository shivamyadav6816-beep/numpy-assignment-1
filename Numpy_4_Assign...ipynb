
# Q9. Vertical Stacking

## Problem Statement
Create two arrays:

```python
[1,2,3]
[4,5,6]
```

Combine them using vertical stacking.

## Sample Output
```python
[[1 2 3]
 [4 5 6]]
```


````python
# Write your code here
arr1 = np.array([[1,2,3]])
arr2 = np.array([[4,5,6]])
np.vstack((arr1,arr2))
````


# Q10. Horizontal Stacking

## Problem Statement
Create two arrays:

```python
[10,20,30]
[40,50,60]
```

Combine them using horizontal stacking.

## Sample Output
```python
[10 20 30 40 50 60]
```


````python
# Write your code here
arr1 = np.array([[10,20,30]])
arr2 = np.array([[40,50,60]])
np.hstack((arr1,arr2))
````
